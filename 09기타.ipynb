```python
n=4

import math
def primenum(n):
    if n==1:
        return False
    for num in range(2,int(math.sqrt(n)+1)):
        if n%num==0:
            return False
    return True
    pass
    
primenum(n)
```

## 1. 에라토스테네스의 체 알고리즘

```python
Image(path+"02.png",width = 600)
```

```python
n=30

array=[i for i in range(2,n+1)]
print(array)

idx=0
while idx<len(array):
    #print('idx: ',idx)
    target=array[idx]
    #print('target: ',target)
    removeList=[x for x in array[idx+1:] if x%target==0]
    #print('removeList: ',removeList)
    array=[x for x in array if x not in removeList]
    #print('array: ',array)
    idx+=1
print(array)
```

### 정석 풀이

```python
n=1000

array=[True for i in range(n+1)]

for i in range(2,int(math.sqrt(n))+1):
    if array[i]==True:
        j=2 #배수 담당
        while (i*j)<=n:
            array[i*j]=False
            j+=1
for i in range(2,n+1):
    if array[i]:
        print(i,end=' ')
```

```python
n,m = 5,5
array=[1,2,3,2,5]

count=0
end=0
total=0
for start in range(n):
    #print('start : ', start)
    while total<m and end<n: #end를 최대까지 이동
        #print('end : ', end)
        #print(array[end])
        total+=array[end]
        #print('total: ',total)
        end+=1
    #print('------------------')
    if total==m:
        count+=1
    total-=array[start] #중복연산 방지 -> 다음 start 오른쪽이동
    #print(array[start])
    #print('total: ',total)
    #print('==================')
    
print(count)
```
