```python
# 자료형 - 사전: 키와 값(쌍), 순서 없음, 키 변경 안됨(immutable), 데이터 조회 수정 O(1)
# 사전을 해쉬테이블이라고 함

# 사전 정의 1번
data = dict()
data['사과'] = 'Apple'
data['바나나'] = 'Banana'
data['코코넛'] = 'Coconut'

print(data)

# 사전 정의 2번
data2 = {
        '나는' : 3,
        '바보' : 4,
        '나는' : 2, # 키 중복 불가
        '바보' : 2 # 값 중복 가능
}
print(data2)

# 특정 키 찾기
if '사과' in data:
    print("\"사과\"가 키로 있음")

# 형변환 필요
print(list(data.keys())) #모든 키 한번에
print(list(data.values())) #모든 값 한번에

# 사전 하나씩 값만 출력
for key in data.keys():
    print(data[key])
    
for v in data.values():
    print(v)
```

### 자료형 - 집합
- 합집합 : a와 b를 모두 합치기
- 교집합 : a와 b의 공통 원소
- 차집합 : a 원소 중 b에 속하지 않는 원소

```python
a = {1,2,3,4,4,5}
b = {5,5,6,7,4,8}

print(a)
print(b)
print(a|b) #합집합
print(a&b) #교집합
print(a-b) #차집합

print()
print(a)
a.add(6) # 한 개 추가
print(a)
a.update([7,8]) #여러개 추가, 리스트 주의
print(a)
a.remove(3) #원소 삭제
print(a)
```

### 함수
- 매개변수 순서 변경 가능 b=1, a=2
- 여러개의 반환값

```python
# global - 지역 변수 없이 함수 바깥에 선언된 변수 참조
a = 0

def func():
    global a
    a += 1

for i in range(10):
    func()

print(a)
```

```python
a = 1

def func():
    a = 0
    a += 1 #UnboundLocalError: a 변수가 선언되지 않은 상태, 초기화 안됨

for i in range(10):
    func()

print(a) # 위에있는 a 참조

# 위에 있는 a와 함수 내 a는 다름
```

```python
# 순열
from itertools import permutations 
data = ['A', 'B', 'C']
result = list(permutations(data, 3))
print(result)
```

```python
# 조합
from itertools import combinations 
data = ['A', 'B', 'C']
result = list(combinations(data, 2))
print(result)
```

```python
# 중복 순열
from itertools import product 
data = ['A', 'B', 'C']
result = list(product(data, repeat=2))
print(result)
```

```python
# 중복 조합
from itertools import combinations_with_replacement 
data = ['A', 'B', 'C']
result = list(combinations_with_replacement(data, 2))
print(result)
```

```python
# 최대 공약수, 최소 공배수
import math

def lcm(a,b): #최소 공배수
    return a*b // math.gcd(a,b)
    # 두 수 곱하기 후 최대공약수 나눈 몫
    # 그냥 외우기

a = 21
b = 14

print(math.gcd(21, 14)) #최대 공약수
print(lcm(21, 14)) # 최소 공배수
```
